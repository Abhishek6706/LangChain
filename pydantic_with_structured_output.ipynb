```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from pydantic import BaseModel, Field
```

```python
from typing import Literal
```

```python
load_dotenv()
model = ChatOpenAI(model='gpt-4.1', temperature = 0.2)
```

```python
class ReviewAnalyizer(BaseModel):

    topic:str = Field(description='Topic name under which the review is written')
    sentiment: Literal['Positive','Negative','Neutral'] = Field(description='Sentiment of the review')
    score: float = Field(description='Rate the sentiment on the scale of 1 to 10')
```

```python
reviews = """
Very Good Product, but i don't like the product.
"""
```

```python
from langchain_core.prompts import PromptTemplate

template = PromptTemplate(
    template="""
Analysize the sentiment of the {review}, and give me the sentiment and its rating""", input_variables= ['review']
)

prompt = template.invoke({'review':reviews})
```

```python
structured_model = model.with_structured_output(ReviewAnalyizer)
```

```python
result = structured_model.invoke(prompt)
```
