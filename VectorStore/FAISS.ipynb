```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_community.document_loaders import PyPDFLoader
from dotenv import load_dotenv
```

```python
load_dotenv()
embedding_model = OpenAIEmbeddings()
```

```python
data = PyPDFLoader(file_path='Fund Facts - HDFC Mid-Cap Fund_February 26.pdf')
```

```python
documents = data.load()
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size = 300, chunk_overlap = 0, separators=['\n\n', '\n', ' ', ''])
```

```python
splitted_data = splitter.split_documents(documents=documents)
```

```python
from langchain_community.docstore.in_memory import InMemoryDocstore
```

```python
import faiss

embedding_dimension = 1536
embedding_dimension_faiss = faiss.IndexFlatL2(embedding_dimension)
```

```python
vectorstore = FAISS(embedding_function=embedding_model, docstore= InMemoryDocstore(), index = embedding_dimension_faiss, index_to_docstore_id={})
```

```python
vectorstore.add_documents(documents=splitted_data)
```

```python
vectorstore.similarity_search(query='Tell me the AUM of the fund', k = 2)
```

```python
vectorstore.similarity_search_with_score(query='Tell me the AUM of the fund', k = 2)
```

```python
vectorstore.similarity_search_with_score(query='Tell me the AUM of the fund', k = 1)
```
