```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.vectorstores import Chroma
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv
```

```python
document = PyPDFLoader(file_path='Fund Facts - HDFC Mid-Cap Fund_February 26.pdf')
```

```python
loaded_document = document.load()
```

```python
loaded_document
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size = 200, chunk_overlap = 0, separators=['\n\n','\n',' ', ''])
```

```python
splitted_document = splitter.split_documents(documents=loaded_document)
```

```python
splitted_document
```

```python
len(splitted_document)
```

```python
load_dotenv()
embedding_model = OpenAIEmbeddings()
```

```python
vector_store = Chroma(collection_name='document_loader_chroma_db', embedding_function=embedding_model, persist_directory='VectorStore\document_loader_chroma')
```

```python
vector_store.add_documents(documents=splitted_document)
```

```python
vector_store.get(ids=['0496a11a-fb8b-479b-80be-66907cc9f968',
 'b3a7e3ee-a497-4645-ac5f-a4c74efc16f5'], include=['embeddings','metadatas','documents'])
```

```python
query = 'Tell me about the AUM, ROI, Manager of the fund'
```

```python
vector_store.similarity_search(query=query, k=3)
```
