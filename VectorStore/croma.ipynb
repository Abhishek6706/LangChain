```python
data1 = Document(page_content='LangChain is an open-source orchestration framework designed to simplify the development of applications powered by Large Language Models (LLMs). It acts as a modular "toolkit" that allows developers to connect an LLM to external data sources, APIs, and databases. By using "chains," the framework can link a sequence of components—such as a specific prompt template, a model like GPT-4, and an output parser—to create a cohesive workflow. It is the industry standard for building Retrieval-Augmented Generation (RAG) systems, which allow AI to provide answers based on specific, private documents rather than just its general training data.', metadata = {'source':'LangChain Official Documentation'})

data2 = Document(page_content='LangGraph is an extension of LangChain designed to handle complex, "loopy" workflows that require state management. While standard LangChain is often linear, LangGraph allows for the creation of cyclical graphs where an AI agent can revisit previous steps, self-correct its errors, or loop through a research phase until a goal is met. By representing an application as a series of nodes (tasks) and edges (paths), it provides a robust way to build persistent, multi-actor agents that can "remember" the context of a long-running process across multiple interactions.', metadata = {'source':'LangGraph Documentation'})

data3 = Document(page_content='The Generative AI (GenAI) Engineer is a specialized role focused on the application layer of the AI stack. These engineers primarily work with "foundation models" (like GPT-4, Claude, or Llama) to build user-facing products. Their expertise lies in prompt engineering, fine-tuning models for specific tasks, and managing vector databases to reduce "hallucinations." Rather than building a model from scratch, a GenAI Engineer focuses on the integration, reliability, and "user-experience" of the AI, ensuring the output is safe, accurate, and contextually relevant.', metadata = {'source':'ZenML: AI Engineering vs ML Engineering'})

data4 = Document(page_content='An ML Engineer operates as the bridge between data science and software engineering, focusing on the entire lifecycle of a machine learning model. Unlike the GenAI Engineer who often consumes pre-built models via APIs, the ML Engineer is responsible for data preprocessing, choosing model architectures, and training models on large datasets. They also manage MLOps, which involves the infrastructure needed to deploy, monitor, and scale models in production, ensuring that systems remain performant and cost-effective.', metadata = {'source':'Google Cloud: AI vs Machine Learning'})
```

```python
data = [data1,data2,data3,data4]
```

```python
from dotenv import load_dotenv
```

```python
load_dotenv()
embedding_model = OpenAIEmbeddings()
```

```python
vector_store = Chroma(collection_name='chroma_db', embedding_function= embedding_model, persist_directory='VectorStore\chroma_vector')
```

```python
vector_store.add_documents(documents=data)
```

```python
vector_store.get(ids=['aa12a4f4-d72f-47aa-89ff-18e71839d21d',
 'ba818d0b-bc71-47d8-930d-ce5a1821456e'], include=['documents','embeddings','metadatas'])
```

```python
vector_store.similarity_search(query='Tell me about LangChain', k = 2)
```

```python
vector_store.similarity_search_with_relevance_scores(query='Tell me about LangChain', k = 2)
```
