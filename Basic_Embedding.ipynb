```python
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv

load_dotenv()

embedding = OpenAIEmbeddings(model= "text-embedding-3-small", dimensions= 40)
result = embedding.embed_query("I am Abhishek Gupta.")
```

```python
print(result)
```

EMBEDDINGS FOR DOCUMENT

```python
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv

load_dotenv()

docs = [
    "My name is Abhishek Gupta",
    "I am from UP",
    "I am a boy",
    "I will become rich one day"
]
embedding = OpenAIEmbeddings(model= "text-embedding-3-small", dimensions= 40)
```

```python
result = embedding.embed_documents(docs, chunk_size= 20)
```

```python
print(result)
```
