```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
from langchain_core.runnables import RunnableParallel, RunnableBranch, RunnableSequence, RunnablePassthrough, RunnableLambda
```

```python
load_dotenv()
model = ChatOpenAI(model='gpt-4.1', temperature=0.2, max_completion_tokens=1000)
```

```python
prompt1 = PromptTemplate(
    template='Generate a post for me for linkedin in a well structured way for {topic}',
    input_variables=['topic']
)
```

```python
prompt2 = PromptTemplate(
    template='Generate a post for me for twitter (now X) in a well structured way for {topic}',
    input_variables=['topic']
)
```

```python
parser = StrOutputParser()
```

```python
parallel_chain = RunnableParallel({
    'LinkedIn': RunnableSequence(prompt1, model, parser),
    'Twitter': RunnableSequence(prompt2, model, parser)
}
)
```

```python
partial_result = parallel_chain.invoke({'topic': 'LangChain VS LangGraph'})
```

```python
partial_result
```

```python
conditional_chain = RunnableBranch(
    (lambda x: (len(x['LinkedIn'])> 100) and (len(x['Twitter'])>100), RunnablePassthrough()),
    RunnableLambda(lambda x: 'Not sufficient data')
)
```

```python
final_chain = parallel_chain | conditional_chain
```

```python
final_result = final_chain.invoke({'topic': 'AI and Antropic'})
```
