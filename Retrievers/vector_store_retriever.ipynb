```python
from langchain_community.vectorstores import Chroma
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv
```

```python
load_dotenv()
embedding_model = OpenAIEmbeddings()
```

```python
document = PyPDFLoader(file_path='Fund Facts - HDFC Mid-Cap Fund_February 26.pdf')
```

```python
document = document.load()
```

```python
document
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size = 250, chunk_overlap = 0, separators=['\n\n','\n', ' ', ''])
```

```python
splitted_document = splitter.split_documents(documents=document)
```

```python
splitted_document
```

```python
len(splitted_document)
```

```python
vector_store = Chroma(collection_name='RAG_Mutual_fund', embedding_function=embedding_model, persist_directory='LangChain\Retrievers\HDFC_Mid_Cap_Fund')
```

```python
vector_store.add_documents(documents=splitted_document)
```

```python
retriever = vector_store.as_retriever(k = 5, fetch_k = 10)
```

```python
query = 'Tell me about the AUM and fund manager of the fund, also give me the historical return of the fund'
```

```python
retrieved_related_document = retriever.invoke(input=query)
```

```python
retrieved_related_document
```

```python
relavant_document = []
for i in range(0,len(retrieved_related_document)):
    relavant_document.append(retrieved_related_document[i].page_content)
```

```python
relavant_document
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
load_dotenv()
model = ChatOpenAI(model = 'gpt-4.1', temperature=0, max_completion_tokens=300)
```

```python
parser = StrOutputParser()
```

```python
template = PromptTemplate(
    template='Give me the response of the query below \n {query} \n\n Basedon the retrieved relelant data which is provided below \n {relavant_document}.',
    input_variables=['query', 'relavant_document']
)
```

```python
chain = template | model | parser
```

```python
chain.invoke({'query':query, 'relavant_document':relavant_document})
```
