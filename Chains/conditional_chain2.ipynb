```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.runnables import RunnableLambda, RunnableSequence, RunnablePassthrough, RunnableBranch, RunnableParallel
from dotenv import load_dotenv
```

```python
load_dotenv()
model = ChatOpenAI(model="gpt-4.1", temperature=0.2, max_completion_tokens= 1000)
```

```python
from langchain_core.prompts import PromptTemplate
```

```python
prompt1 = PromptTemplate(
    template='Generate the {topic} in such a way so that i can use it for my {use_case}.',
    input_variables=['topic', 'use_case']
)
```

```python
from langchain_core.output_parsers import StrOutputParser
```

```python
parser = StrOutputParser()
```

```python
prompt2 = PromptTemplate(
    template='Generate top 10 questions that can be asked from the below {topic} for the {use_case}.',
    input_variables=['topic', 'use_case']
)
```

```python
chain = RunnableParallel({
    'topic': RunnableSequence(prompt1, model, parser),
    'use_case' : RunnableSequence(prompt2, model, parser)
})
```

```python
chain_branch = RunnableBranch(
    (lambda x: chain is not None, chain),  
    RunnableLambda(lambda x: "The topic is too broad, try to be more specific")
)
```

```python
final_chain = RunnableSequence(chain, chain_branch)
```

```python
result = final_chain.invoke({'topic': 'Artificial Intelligence', 'use_case': 'Interview preparation'})
```
