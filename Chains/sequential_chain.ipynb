```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from dotenv import load_dotenv
from langchain_core.output_parsers import StrOutputParser
```

```python
load_dotenv()
model = ChatOpenAI(model='gpt-4.1', temperature=0.2, max_completion_tokens=3000)
```

```python
template1 = PromptTemplate(
    template='Generate a detail story on {name}',
    input_variables=['name']
)
```

```python
template2 = PromptTemplate(
    template='Extract the top 5 features from the below text /n {text}',
    input_variables=['text']
)
```

```python
parser = StrOutputParser()
```

```python
chain = template1 | model | parser | template2 | model | parser
```

```python
result = chain.invoke({'name':'Virat Kohli'})
```
