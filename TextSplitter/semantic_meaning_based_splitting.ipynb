```python
from langchain_experimental.text_splitter import SemanticChunker
from langchain_openai import OpenAIEmbeddings
```

```python
from dotenv import load_dotenv
load_dotenv()
embedding_model = OpenAIEmbeddings()
```

```python
text = '''As war escalates in the Middle East and disrupts global supply chains, the ripple effects are fundamentally rewiring the financial landscape and daily life. The stock market is currently buckling under the weight of stagflation fears, with major indices tumbling as surging energy prices force central banks to delay expected interest rate cuts. In this climate of severe uncertainty, investors have flooded into safe-haven assets, driving gold to unprecedented highs above $5,000 an ounce while throwing the broader commodities sector into volatile swings. Caught in the crossfire of these macroeconomic shifts are farmers, who are struggling to operate essential agricultural machinery amidst skyrocketing fuel costs and supply shortages. Ultimately, these compounding crises trickle down to everyday people, who are left to bear the brunt of the resulting inflation and threatened food supplies at the grocery store.'''
```

```python
text_splitter = SemanticChunker(embeddings=embedding_model)
```

```python
text_splitter.create_documents([text])
```
