```python
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv
from sklearn.metrics.pairwise import cosine_similarity
```

```python
load_dotenv()

embeddings = OpenAIEmbeddings(model = "text-embedding-3-large", dimensions = 1000)
```

```python
documents = [
             "Kellton Tech Solution company is an IT company with 1000 cr valuation.",
             "Infosys IT company have the valuation of 5 lakh crores.",
             "TCS is an Tata IT firm having 10 lakh c rores as its valuation",
             "Anthropic is an American IT company which comes in 2021 has the valuation of 35 lakh crores."
             ]
```

```python
query = "Tell me about Anthropic"
```

```python
query_embeddings = embeddings.embed_query(query)
```

```python
document_embeddings = embeddings.embed_documents(documents)
```

```python
similarity = cosine_similarity([query_embeddings], document_embeddings)
```

```python
similarity
```

```python
most_similar_index, similarity_score = sorted(list(enumerate(similarity[0])))[-1]
```

```python
print(query)
print(documents[most_similar_index])
print(similarity_score)
```
