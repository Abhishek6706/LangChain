```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_community.document_loaders import CSVLoader
from dotenv import load_dotenv
from langchain_core.output_parsers import StrOutputParser
```

```python
load_dotenv()
llm = ChatOpenAI(model='gpt-4.1', temperature=0.2, max_completion_tokens=1000)
```

```python
loader = CSVLoader(file_path='top10nifty50_120326.csv', autodetect_encoding=True)
```

```python
docs = loader.load()
```

```python
data = '''

'''
```

```python
for i in range(0, len(docs)):
    data = data + docs[i].page_content
```

```python
data
```

```python
prompt = PromptTemplate(
    template='Explain the below data \n {data}',
    input_variables=['data']
)
```

```python
parser = StrOutputParser()
```

```python
chain = prompt | llm | parser
```

```python
result = chain.invoke({'data':data})
```
