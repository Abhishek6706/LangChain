```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_community.document_loaders import PyPDFLoader
from dotenv import load_dotenv
```

```python
load_dotenv()
llm = ChatOpenAI(model='gpt-4.1', temperature=0.2, max_completion_tokens=1000)
```

```python
loader = PyPDFLoader(file_path='Fund Facts - HDFC Mid-Cap Fund_February 26.pdf')
```

```python
docs_py_pdf = loader.load()
```

```python
docs_py_pdf
```

```python
docs_py_pdf[0].page_content
```

```python
prompt = PromptTemplate(
    template='give me top 3 important information from the below text \n {text}',
    input_variables=['text']
)
```

```python
parser = StrOutputParser()
```

```python
chain = prompt | llm | parser
```

```python
text = '''

'''

for i in range(0, len(docs_py_pdf)):
    text = text + docs_py_pdf[i].page_content
```

```python
result = chain.invoke({'text': text})
```
