```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from dotenv import load_dotenv
from langchain_core.output_parsers import StrOutputParser
from langchain_community.document_loaders import TextLoader
```

```python
load_dotenv()
llm = ChatOpenAI(model='gpt-4.1', temperature=0.2, max_completion_tokens= 1000)
```

```python
loader = TextLoader(file_path='mutual_fund.txt', autodetect_encoding=True)
```

```python
docs = loader.load()
```

```python
docs
```

```python
len(docs)
```

```python
prompt = PromptTemplate(
    template='Give me the top 3 key points from the below text \n {text}',
    input_variables=['text']
)
```

```python
parser = StrOutputParser()
```

```python
chain = prompt| llm | parser
```

```python
result = chain.invoke({'text': docs[0].page_content})
```
