```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
from typing import TypedDict, Annotated
```

```python
review_text = """
It’s the best model in the world for coding, agents, and computer use. It’s also meaningfully better at everyday tasks like working with slides and spreadsheets. When we gave it our two-hour engineering assignment, it finished faster than any human ever has.
"""
```

```python
class review(TypedDict):
    review: Annotated[str, 'Topic name under which the review is written']
    sentiment: Annotated[str, 'sentiment must be either positive, negative, neutral']
    score: Annotated[float, 'Score in between 1-10']
```

```python
load_dotenv()
model = ChatOpenAI(model='gpt-4.1', temperature=0.2)
```

```python
structured_model = model.with_structured_output(review)
```

```python
from langchain_core.prompts import PromptTemplate
```

```python
prompt = PromptTemplate(
    template="""
Generate the sentiment and its score on the scale of 1-10, for the below review
{review_text}

Don't hallucinate.,
""", input_variables= ['review_text']
)
```

```python
final_prompt = prompt.invoke({'review_text':review_text})
```

```python
result = structured_model.invoke(final_prompt)
```
