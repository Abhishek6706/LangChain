```python
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import PydanticOutputParser
from dotenv import load_dotenv
from pydantic import BaseModel, Field
from langchain_core.prompts import PromptTemplate
```

```python
load_dotenv()
model = ChatOpenAI(model='gpt-4.1', temperature=0.2, max_completion_tokens=300)
```

```python
from typing import Literal, List
```

```python
class movie_character(BaseModel):
    name: str = Field(description='Character name in the movie')
    species: str = Field(description='Species of the character such as human, witch, wizard, etc')
    sex: Literal['male','female','others'] = Field(description='Gender of the character')
```

```python
class CharacterList(BaseModel):
    characters: List[movie_character] = Field(description="A list of characters from the movie")
```

```python
parser = PydanticOutputParser(pydantic_object=CharacterList)
```

```python
template = PromptTemplate(
    template = 'Give me 10 characters which is present in the movie, {movie} \n {format_instruction}',
    input_variables=['movie'],
    partial_variables={'format_instruction':parser.get_format_instructions()}
)
```

```python
chain = template | model | parser
```

```python
result = chain.invoke({'movie':'Harry Potter'})
```
